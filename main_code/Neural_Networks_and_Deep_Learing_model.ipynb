```python
# Import dependencies
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
# Read CSV file
df = pd.read_csv("exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()
df.head()
```

# Select features (columns)

```python
selected_features = df
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign x and y values
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]

print(X.shape, y.shape)
```

```python
# Import Dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Split data in training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
X_train.head()
```

# Pre-processing
Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train Model - Neural Network
Use `Neural Network` model

```python
from tensorflow.keras.utils import to_categorical
```

```python
# Step 1: Label encoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
print(label_encoder.classes_)
```

```python
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
# pd.plotting.scatter_matrix(X, c=encoded_y, figsize=(40,40));
```

```python
# First, create a normal neural network
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=15, activation='relu', input_dim=40))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(X_train_scaled, y_train_categorical, epochs=100, shuffle=True, verbose=2)
```

# Deep Learning

```python
deep_model = Sequential()
deep_model.add(Dense(units=15, activation='relu', input_dim=40))
deep_model.add(Dense(units=15, activation='relu'))
deep_model.add(Dense(units=3, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

deep_model.fit(X_train_scaled, y_train_categorical, epochs=100, shuffle=True, verbose=2)
```

# Compare the two models (Neural Networks and Deep Learning)

```python
# Neural Networks
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f'Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}')
```

```python
# Deep Learning
model_loss, model_accuracy = deep_model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f'Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}')
```

# Save the Model

```python
# Saving neural model
model.save('efmercado_normal_neural.h5')

# Saving deep model
deep_model.save('efmercado_deep_learning.h5')
```
